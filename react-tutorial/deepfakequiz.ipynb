```python
f = open("ASV2019.txt")
lines = f.readlines()
asv_vals = [ (100.0-float(line.strip()))/100.0 for line in lines ]
#print(asv_vals)

f = open("grades.txt")
lines = f.readlines()
user_vals = [ float(line.strip().split(",")[3])/20.0 for line in lines ]
#print(user_vals)

dict1 = {'humans': user_vals, 'machines': asv_vals}
labels, data = [*zip(*dict1.items())] 
#print(labels,data)

import matplotlib.pyplot as plt
plt.boxplot(data)
plt.xticks(range(1, len(labels) + 1), labels)
plt.suptitle("Detecting Deep Fake Speech - Humans vs Machines")
plt.title("ASVSpoof 2019 DataSet")

fig = plt.gcf()
fig.savefig("asvspoof_compare.png")
plt.show()

```
